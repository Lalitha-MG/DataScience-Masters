Q3. Write a Python function to calculate the probability density function of a normal distribution with 
given mean and standard deviation at a given point

A normal distribution is a type of continuous probability distribution for a real-valued random variable. It is based on mean and standard deviation. The probability distribution function or PDF computes the likelihood of a single point in the distribution. 

We will use scipy.stats.norm.pdf() method to calculate the probability distribution for a number x.

Syntax: scipy.stats.norm.pdf(x, loc=None, scale=None)

Parameter:

x : array-like object, for which probability is to be calculated.
loc : optional (default=0), represents mean of the distribution.
scale : optional (default=1), represents standard deviation of the distribution.
Returns: A probability density function calculated at x as a ndarray object.

```python
from scipy.stats import norm
import numpy as np

data_start = -5
data_end = 5
data_points = 11
data = np.linspace(data_start, data_end, data_points)

mean = np.mean(data)
std = np.std(data)

probability_pdf = norm.pdf(3, loc=mean, scale=std)
print(probability_pdf)
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

data_start = -3
data_end = 3
data_points=7

data=np.linspace(data_start,data_end,data_points,dtype=int)
print(data)

mean=np.mean(data)
print(mean)

stddev=np.std(data)
print(stddev)

probability_pdf=norm.pdf(2,loc=mean,scale=stddev)
print(probability_pdf)

plt.plot(data, norm.pdf(data, mean, stddev))
plt.show()
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial 
distribution can be applied.

The binomial distribution describes the behavior of a count variable X if the following conditions apply:

1: The number of observations n is fixed.

2: Each observation is independent.

3: Each observation represents one of two outcomes ("success" or "failure").

4: The probability of "success" p is the same for each outcome.

Ex : 1) Probability of having 3 Heads when coin is tossed 10 times

2) Probability of 2 scores in 5 attempts

```python
#PMF for Poisson Distribution
#Import the dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
#we will need an array of the k values for which we will compute the Poisson PMF.
k = np.arange(0, 17)
print(k)
#In order to calculate the Poisson PMF using Python, we will use the .pmf() method of the scipy.poisson generator. It will need two parameters:
#k value (the k array that we created)
# value (which we will set to 7 as in our example)
pmf = poisson.pmf(k, mu=7)
pmf = np.round(pmf, 5)
print(pmf)
for val, prob in zip(k,pmf):
    print(f"k-value {val} has probability = {prob}")
#Using matplotlib library, we can easily plot the Poisson PMF using Python:
plt.plot(k, pmf, marker='o')
plt.xlabel('k')
plt.ylabel('Probability')
plt.title('Poisson Probability Distribution using PMF')
plt.show()
```

Q7. How Binomial distribution different from Poisson distribution

Binomial Distribution provides the number of successes after n trials where probability of success of each trail is p

Poisson Distribution predicts the probability of agiven no of events occuring in a fixed interval of time

Binomial Distribution	
1) It is biparametric, i.e. it has 2 parameters n and p.
2) The number of attempts are fixed.
3) The probability of success is constant.
4) There are only two possible outcomes-Success or failure.
5) Mean>Variance

Poisson Distribution
1) It is uniparametric, i.e. it has only 1 parameter m.
2) The number of attempts are unlimited.
3) The probability of success is extremely small.
4) There are unlimited possible outcomes.
5) Mean=Variance
